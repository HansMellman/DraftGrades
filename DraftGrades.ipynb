```python
# Isolate only needed columns from player_information csv.
player_information = player_information[['Player_ID', 'First_Name', 'Last_Name', 'Position', 'Drafted_By', 'Draft_Round', 'Drafted_Position', 'Draft_Year']]
# Isolate only needed columns from players_personal csv.
players_personal = players_personal[['Player_ID', 'Future_Overall']]
# Isolate only needed columns from team_information csv.
team_information = team_information[['Team', 'Home_City']]
```

```python
# Isolate only entries for current draft year.
# Filter the DataFrame for the current year and remove entries with 'P' or 'K' in the Position column
cur_draft = player_information[
    (player_information['Draft_Year'] == cur_year) &
    (~player_information['Position'].str.contains('P|K', na=False))
].copy()

# Step 1: Create the mapping dictionary from team_information DataFrame
team_mapping = team_information.set_index('Team')['Home_City'].to_dict()

# Step 2: Overwrite the 'Drafted_By' column with team names in cur_draft
# cur_draft.loc[:, 'Drafted_By'] = cur_draft['Drafted_By'].map(team_mapping)
cur_draft['Drafted_By'] = cur_draft['Drafted_By'].map(team_mapping)
# merge in for info from players_personal
cur_draft = pd.merge(cur_draft, players_personal, how='left', on='Player_ID')
```

```python
# Step 1: Load the EstimatedFV CSV into a DataFrame
estimated_fv_df = pd.read_csv('EstimatedFV.csv')

# Step 2: Merge the DataFrames, keeping only the 'expected_FV' column
cur_draft = pd.merge(cur_draft, 
                     estimated_fv_df[['draft round', 'draft_position', 'expected_FV']], 
                     how='left', 
                     left_on=['Draft_Round', 'Drafted_Position'], 
                     right_on=['draft round', 'draft_position'])

# Step 3: Drop the extra 'draft round' and 'draft_position' columns that were added during the merge
cur_draft.drop(columns=['draft round', 'draft_position'], inplace=True)

# Step 4: Convert 'expected_FV' column to int64, handling any NaN values
cur_draft['expected_FV'] = cur_draft['expected_FV'].fillna(0).astype('int64')
```

```python
# Get the count of each unique value in the 'expected_FV' column ## IF NEEDED TO TROUBLESHOOT.
value_counts = cur_draft['expected_FV'].value_counts()

# Find the count of 0's specifically
zeros_count = value_counts.get(0, 0)

print(f"Number of 0's in the 'expected_FV' column: {zeros_count}")
```

```python
cur_draft['difference_BPA'] = (cur_draft['Future_Overall'] - cur_draft['expected_FV'])
```

```python
def get_pos_avg(draft_round, position):
    # Select the appropriate dictionary based on the draft round
    if draft_round == 1:
        return round_1.get(position, 0)
    elif draft_round == 2:
        return round_2.get(position, 0)
    elif draft_round == 3:
        return round_3.get(position, 0)
    elif draft_round == 4:
        return round_4.get(position, 0)
    elif draft_round == 5:
        return round_5.get(position, 0)
    elif draft_round == 6:
        return round_6.get(position, 0)
    elif draft_round == 7:
        return round_7.get(position, 0)
    else:
        return 0  # If the draft round is out of bounds or not available

# Recompute pos_avg
cur_draft['pos_avg'] = cur_draft.apply(lambda row: get_pos_avg(row['Draft_Round'], row['Position']), axis=1)
```

```python
# Calculate the weighted value and apply floor operation
cur_draft['weighted_value'] = np.floor(cur_draft['Future_Overall'] * 
                                       (cur_draft['Future_Overall'] / cur_draft['pos_avg'])).astype(int)
```

```python
# Group by 'Drafted_By' and calculate the mean of 'difference_BPA'
bpa_df = cur_draft.groupby('Drafted_By', as_index=False)['difference_BPA'].mean().round(2)

# Rename the 'difference_BPA' column to 'bpa' to reflect the calculated average
bpa_df.rename(columns={'difference_BPA': 'bpa'}, inplace=True)

# Sort the DataFrame by the 'bpa' column in descending order
bpa_df = bpa_df.sort_values(by='bpa', ascending=False)

# Add a rank column based on the sorted 'bpa' values
bpa_df['bpa_rank'] = bpa_df['bpa'].rank(ascending=False, method='min').astype(int)

# Define the conditions for the 'verdict' column
conditions = [
    (bpa_df['bpa'] > 2.01),
    (bpa_df['bpa'] >= -2.00) & (bpa_df['bpa'] <= 2.00),
    (bpa_df['bpa'] < -2.01)
]

# Define the corresponding verdicts
verdicts = [
    'Did well for BPA',
    'Drafted as expected',
    'Missed on picks or prioritised other things over BPA'
]

# Add the 'verdict' column to the DataFrame
bpa_df['verdict'] = np.select(conditions, verdicts, default='Unknown')

# Reorder columns to have 'rank' as the first column
bpa_df = bpa_df[['bpa_rank', 'Drafted_By', 'bpa', 'verdict']]

# Reset the index of the DataFrame
bpa_df.reset_index(drop=True, inplace=True)

# Display the final DataFrame
bpa_df
```

```python
# Group by 'Drafted_By' and calculate the mean of 'weighted_value'
weighted_df = cur_draft.groupby('Drafted_By', as_index=False)['weighted_value'].mean().round(2)

# Rename the 'difference_BPA' column to 'bpa' to reflect the calculated average
weighted_df.rename(columns={'weighted_average': 'weighted_value'}, inplace=True)

# Sort the DataFrame by the 'bpa' column in descending order
weighted_df = weighted_df.sort_values(by='weighted_value', ascending=False)

# Add a rank column based on the sorted 'bpa' values
weighted_df['weighted_rank'] = weighted_df['weighted_value'].rank(ascending=False, method='min').astype(int)

# Define the conditions for the 'verdict' column
conditions = [
    (weighted_df['weighted_value'] >= 50),
    (weighted_df['weighted_value'] <= 43)
]

# Define the corresponding verdicts
verdicts = [
    'Acquired 2+ Starters',
    'Had Several Misses',
]

# Add the 'verdict' column to the DataFrame
weighted_df['verdict'] = np.select(conditions, verdicts, default='Unknown')

# Reorder columns to have 'rank' as the first column
weighted_df = weighted_df[['weighted_rank', 'Drafted_By', 'weighted_value', 'verdict']]

# Reset the index of the DataFrame
weighted_df.reset_index(drop=True, inplace=True)

# Display the final DataFrame
weighted_df
```

```python
# Merge bpa_df with weighted_df on the 'Drafted_By' column
combined_df = pd.merge(bpa_df[['Drafted_By', 'bpa_rank']], weighted_df, on='Drafted_By', how='left')

# Rename the 'rank' column from bpa_df to 'bpa_rank' to avoid confusion
# combined_df.rename(columns={'rank': 'bpa_rank'}, inplace=True)

# Reorder columns to have 'bpa_rank' as the first column
combined_df = combined_df[['bpa_rank', 'weighted_rank', 'Drafted_By', 'weighted_value', 'verdict']]

# Sort the DataFrame by the 'weighted_rank' column in ascending order
combined_df = combined_df.sort_values(by='weighted_rank', ascending=True)

# Reset the index again after sorting, if desired
combined_df.reset_index(drop=True, inplace=True)

# Display the final sorted DataFrame
combined_df
```

```python
# Calculate the mean of 'difference_BPA' per draft round
draft_quality = cur_draft.groupby('Draft_Round', as_index=False)['difference_BPA'].mean()

# Rename the columns to match the desired format
draft_quality.rename(columns={'difference_BPA': 'average value'}, inplace=True)

# Sort by round to ensure it's in order
draft_quality = draft_quality.sort_values(by='Draft_Round')

# Round the 'average value' column to two decimal places
draft_quality['average value'] = draft_quality['average value'].round(2)

# Reset the index (optional, for clean indexing)
draft_quality.reset_index(drop=True, inplace=True)

# Optionally, add an 'overall' row to represent the overall average value across all rounds
overall_mean = cur_draft['difference_BPA'].mean().round(2)
draft_quality.loc[len(draft_quality.index)] = ['overall', overall_mean]

# Display the final DataFrame
draft_quality
```

```python
## Write the dataframes to an excel file to be shared/saved.
def save_dfs_to_excel(**dfs):
    filename = f"{league}_{cur_year}_DraftGrades.xlsx"
    
    # Create an Excel writer object with the filename based on the league_cur_year
    with pd.ExcelWriter(filename, engine='xlsxwriter') as writer:
        # Iterate over the dictionary of DataFrames and their sheet names
        for sheet_name, df in dfs.items():
            # Write the DataFrame to the corresponding sheet
            df.to_excel(writer, sheet_name=sheet_name, index=False)

# Example usage with your DataFrames
save_dfs_to_excel(bpa_df=bpa_df, 
                  combined_df=combined_df, 
                  draft_quality=draft_quality)
```
